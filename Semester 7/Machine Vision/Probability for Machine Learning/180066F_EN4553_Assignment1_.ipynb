```python
import numpy as np
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
mu_1 = 0
mu_2 = 10
sigma_1 = 1
sigma_2 = 1
p = 0.2

x1 = sigma_1 * np.random.randn(1,10000) + mu_1  #PDF of X1
x2 = sigma_2 * np.random.randn(1,10000) + mu_2  #PDF of X2

I=np.random.binomial(n=1, p=p, size=(100,1))  #Binomial RV

y = I*x1+(1-I)*x2                             #defined Y
```

```python
sns.distplot(x1)
```

```python
sns.distplot(x2)
```

```python
sns.distplot(y, color ='red')
```
